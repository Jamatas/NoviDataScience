# Aantal indicators per jaar

```python
df_Indicators_per_Year = df_Indicators['IndicatorCode'].groupby(df_Indicators['Year']).count()
df_Indicators_per_Year
```

```python
plt.plot(df_Indicators_per_Year)
```

# Mobile subscriptions

```python
df_Series[df_Series['IndicatorName'].str.contains('Mobile')]
```

```python
df_MobileSubscriptions = df_Indicators[(df_Indicators['IndicatorCode']=='IT.CEL.SETS')]
df_MobileSubscriptions.head()
```

```python
df_Euro = df_MobileSubscriptions[df_MobileSubscriptions['CountryCode']=='EMU']
df_EastAsia = df_MobileSubscriptions[df_MobileSubscriptions['CountryCode']=='EAS']

plt.plot(df_Euro['Year'], df_Euro['Value'], marker='o')
plt.plot(df_EastAsia['Year'], df_EastAsia['Value'], marker='+')
plt.show()
```
