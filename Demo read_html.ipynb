# Data import

```python
import pandas as pd
```

```python
url = 'https://www.indexmundi.com/map/?v=25&r=xx&l=nl'
table_id = 'gvDataCountry'

list_of_dfs  = pd.read_html(url, attrs={"id":table_id})

df_raw = list_of_dfs[0]
headers = df_raw.iloc[0]
df  = pd.DataFrame(df_raw.values[1:], columns=headers)
df
```
