```python
df = pd.read_csv('datasets-uci-iris.csv', 
                 index_col=False,
                 names=['sepal length (cm)', 
                        'sepal width (cm)', 
                        'petal length (cm)', 
                        'petal width (cm)',
                        'target'])
```

```python
print('%d rows x %d columns' % df.shape)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,3))

for i, title in enumerate(df.columns[:-1], 1):
    plt.subplot(1, 4, i)
    plt.title(title)
    df[title].hist()

plt.show()
```

```python
targets = df['target'].unique()
targets
```

```python
color_mapping = dict(zip(targets, ('red','green','blue')))
color_mapping
```

```python
colors = df['target'].map(color_mapping)
```

```python
sc = pd.plotting.scatter_matrix(df, 
                                alpha=0.3, 
                                figsize=(12, 12), 
                                diagonal='hist', 
                                color=colors, 
                                marker='o', 
                                grid=True,
                               )
```
